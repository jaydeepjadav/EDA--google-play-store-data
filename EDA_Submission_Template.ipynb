````python
#converting column 'Installs'  into float datatype

df1['Installs'] = df1['Installs'].apply(lambda x: str(x).replace('+',''))    
df1['Installs'] = df1['Installs'].apply(lambda x: str(x).replace(',',''))
df1['Installs'] = df1['Installs'].apply(lambda x: float(x))
````

````python
#for column 'Price'
df1['Price'] = df1['Price'].apply(lambda x: str(x).replace('$',''))
df1['Price'] = df1['Price'].apply(lambda x: float(x))
````

````python
df1['Reviews'] =pd.to_numeric(df1['Reviews'],errors='coerce') 
# Using errors='coerce'. It will replace all non-numeric values with NaN.
````
